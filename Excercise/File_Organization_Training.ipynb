# Python Training: Automating File Organization 📁

Welcome to the file organization training module! In this session, we will learn how to use Python to clean up digital clutter. 

### The Real-Life Scenario:
Imagine your **Downloads** folder is an absolute disaster. It's filled with PDFs, images, ZIP files, and spreadsheets all mixed together. Doing this manually takes forever. We are going to build a Python script that automatically scans a folder and sorts files into dedicated subfolders based on their file extensions.

### What you will learn:
1. How to navigate the file system using `os`.
2. How to safely move and copy files using `shutil`.
3. How to write a robust automation script.

```python
# Setup Cell: Let's create a mock "Messy_Downloads" folder for testing
import os

# Define our target directory
target_dir = "Messy_Downloads"

# Create the folder if it doesn't exist
os.makedirs(target_dir, exist_ok=True)

# Create some dummy files to simulate real-life clutter
dummy_files = [
    "resume.pdf", "invoice_1024.pdf", "tutorial.pdf",
    "vacation_pic.jpg", "profile_baked.png", "logo.svg",
    "budget_2026.xlsx", "quarterly_report.csv",
    "project_files.zip", "backup.tar.gz"
]

for file in dummy_files:
    file_path = os.path.join(target_dir, file)
    # Create an empty file
    with open(file_path, 'w') as f:
        f.write("Dummy content for training purposes.")

print(f"✅ Setup complete! Created '{target_dir}' folder with {len(dummy_files)} messy files.")
```

## Step 1: Scanning the Folder

Before we can organize anything, we need to see what is inside the folder. We will use `os.listdir()` to grab a list of all files.

```python
import os

target_dir = "Messy_Downloads"

# List everything inside the directory
all_items = os.listdir(target_dir)

print("Files found in the messy folder:")
for item in all_items:
    print(f"- {item}")
```

## Step 2: Extracting File Extensions

To sort files, we need to know what *type* of file they are. We can extract the extension (like `.pdf` or `.jpg`) using `os.path.splitext()`.

Let's test this concept.

```python
# Testing extension extraction on a single filename
filename = "budget_2026.xlsx"

# splitext splits the path into a tuple: (name, extension)
name, extension = os.path.splitext(filename)

print(f"Original: {filename}")
print(f"Name part: {name}")
print(f"Extension part: {extension}")
```

## Step 3: Mapping Extensions to Folders

Now, let's define a dictionary that decides where each file type should go. If an extension isn't in our dictionary, we'll toss it into an "Others" folder.

```python
# Define our folder mapping logic
extension_map = {
    '.pdf': 'Documents',
    '.xlsx': 'Spreadsheets',
    '.csv': 'Spreadsheets',
    '.jpg': 'Images',
    '.png': 'Images',
    '.svg': 'Images',
    '.zip': 'Archives',
    '.gz': 'Archives'
}

print("Mapping logic defined successfully!")
```

## Step 4: Putting it Together (The Automation Script)

We will now combine everything. We will use `shutil.move()` to physically relocate the files.

### The Logic Flow:
1. Loop through each file in `Messy_Downloads`.
2. Get its extension.
3. Determine the destination folder based on our map.
4. Create the destination folder if it doesn't exist yet (`os.makedirs`).
5. Move the file.

```python
import os
import shutil

source_folder = "Messy_Downloads"

# 1. Loop through the files
for filename in os.listdir(source_folder):
    source_path = os.path.join(source_folder, filename)
    
    # Make sure we are only moving files, not folders
    if os.path.isfile(source_path):
        # 2. Get extension
        _, ext = os.path.splitext(filename)
        ext = ext.lower() # Normalize to lowercase
        
        # 3. Determine destination folder name
        folder_name = extension_map.get(ext, 'Others')
        destination_folder = os.path.join(source_folder, folder_name)
        
        # 4. Create the subfolder if it doesn't exist
        os.makedirs(destination_folder, exist_ok=True)
        
        # 5. Move the file
        destination_path = os.path.join(destination_folder, filename)
        shutil.move(source_path, destination_path)
        print(f"Moved: {filename} -> {folder_name}/")

print("\n🎉 File organization complete!")
```

## Step 5: Verify the Results

Let's inspect the folder structure now to see if our Python script successfully cleaned up the mess.

```python
print(f"Current structure of '{source_folder}':\n")

for root, dirs, files in os.walk(source_folder):
    # This helps format a nice visual tree structure
    level = root.replace(source_folder, '').count(os.sep)
    indent = ' ' * 4 * (level)
    print(f"{indent}[Folder] {os.path.basename(root) or source_folder}")
    subindent = ' ' * 4 * (level + 1)
    for file in files:
        print(f"{subindent}{file}")
```
